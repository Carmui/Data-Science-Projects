```python
import pandas as pd # data processing 
import numpy as np # linear algebra
import matplotlib as plt # data visualisation

%matplotlib inline
```

```python
# read the data
pokemon = pd.read_csv("pokemon.csv", index_col = "#")
pokemon.head()
```

```python
# Checking out the data type, number of columns, number of rows, memory usage
pokemon.info(memory_usage = 'deep')
```

```python
# Checking out the columns names
pokemon.columns
```

```python
# Head shows the first 10 rows
pokemon.head(10)
```

```python
# In contrast to head we can use tail to show last 10 rows
pokemon.tail(10)
```

```python
# We can use value_counts() function to see how many legendary pokemons are there...
pokemon["Legendary"].value_counts()
```

# Filtering Data

```python
# We can find out what are the names of pokemon types
pokemon["Type 1"].unique()
```

```python
# Let's filter the legendary pokemons only! (first 5 of them)
legendary_filter = pokemon["Legendary"] == True
legendary = pokemon[legendary_filter]
legendary.head()
```

```python
# What if we would like to show only the strongest pokemons? (first 5 of them)
attack_filter = pokemon["Attack"] > 150
strongest = pokemon[attack_filter]
strongest.head()
```

```python
# Lets make some advanced filtering - get the strongest legendary pokemons! 
pokemon[legendary_filter & attack_filter]
```

```python
# We can show pokemons with Speed = 150, using parametr 'all' since we can find out some of the pokemons have NaN in Type2!
speed = pokemon.where(pokemon["Speed"] == 150)
speed.dropna(how = 'all')
```

```python
# To filter data we can also use 'query' statement 
# for example to search for pokemons that have less than 20 defense and more than 50 attack
pokemon.query('Attack >= 50 and Defense <= 20')
```

# Groupby 


```python
# We can filter data by groupby as well
grouped_pokemon = pokemon.groupby(by = 'Type 1')
grouped_pokemon.groups
```

```python
# By initializing grouped_pokemon which is grouped by Type1, we can now easily filter f.e. type dragon!
grouped_pokemon.get_group('Dragon').head()
```

# Multiindexing

```python
# Firstly, lets reset the index
pokemon.reset_index(inplace = True)
pokemon.head()
```

```python
# Lets set new index based on two columns
pokemon.set_index(keys = ["Name", "Generation"],inplace = True)
pokemon.head()
```

```python
# Hmm, what if we swap index level now?
pokemon = pokemon.swaplevel(i = 0, j = 1)
pokemon.head()
```

```python
# We can quite easy get full pokemon list for each generation
pokemon.loc[3].head(10)
```

```python
# Actually, the multiindexing isn't necessary in this case, lets swap it back
pokemon.reset_index(inplace = True)
pokemon.set_index('#', inplace = True)
pokemon.head()
```

# Pivot tables

```python
# I decided to create pivot table to check what is the average attack and defense for some of the Type 1 types
# I dropped NaN values to check only the most popular Type 1 types
piv = pokemon.pivot_table(index = "Type 1", columns = "Generation", values = ["Attack", "Defense"], aggfunc = 'mean').dropna()
piv
```

```python
# Unpivoting table by using melt
piv.reset_index(inplace = True)
piv.melt(id_vars = "Type 1")
```

# Data creation & modification

```python
# I decided to add an additional column named 'Mixed-Type' which contains both Types in one 
# if the pokemon has one type, NaN is fulfilled by the Type 1
pokemon['Mixed-Type'] = pokemon['Type 1'] + ' ' + pokemon['Type 2']
pokemon['Mixed-Type'].fillna(value = pokemon['Type 1'], inplace = True)
pokemon.head()
```

```python
# For the next steps lets delete the Mixed-Type column
pokemon.drop(inplace = True, columns = 'Mixed-Type' )
```

# Functions & lambdas

```python
def ChangeLegendary(word):
    if word == False:
        return 'Normal'
    else: 
        return 'Legendary'
```

```python
# Lets check if function work properly
legend_pokemons = pokemon['Legendary'] 
legend_pokemons = legend_pokemons.apply(ChangeLegendary)
legend_pokemons
```

```python
legend_pokemons = legend_pokemons.apply(lambda x: str(x).upper())
legend_pokemons
```

# Plots

#### 1. Line plot

```python
fig, (ax1, ax2) = plt.pyplot.subplots(figsize = (15,10), nrows = 2, ncols = 1, sharex = True, )
pokemon['Attack'].plot(ax = ax1, color = 'r', title = 'Attack', linewidth = 0.75)
ax1.set_ylabel("Atk")
ax1.legend()

pokemon['Sp. Atk'].plot(ax = ax2, color = 'b', title = 'Sp. Atk', linewidth = 1, linestyle ='dotted')
ax2.set_ylabel("Sp.Atk")
ax2.legend()

plt.pyplot.xlabel('Pokemons')
```

#### 2. Bar plot

```python
figure = plt.pyplot.subplots(1, figsize = (15, 10))
plt.pyplot.subplot(211)
plt.pyplot.bar(pokemon["Type 1"], pokemon["Defense"], width=.7)
plt.pyplot.xlabel('Type')
plt.pyplot.ylabel('Defense')

plt.pyplot.subplot(212)
plt.pyplot.bar(pokemon["Type 1"], pokemon["HP"], color = "r", width=.7)
plt.pyplot.xlabel('Type')
plt.pyplot.ylabel('HP')
```

#### 3. Histogram Plot

```python
plt.pyplot.subplots(figsize = (15, 10))
plt.pyplot.subplot(211)
plt.pyplot.hist(pokemon["Sp. Atk"], bins = 40)
plt.pyplot.ylabel("Count")
plt.pyplot.xlabel("Sp. Atk value")
plt.pyplot.subplot(212)
plt.pyplot.hist(pokemon["Sp. Atk"], bins = 40, color = 'Green', orientation = 'horizontal')
plt.pyplot.ylabel("Sp. Atk value")
plt.pyplot.xlabel("Count")
```

#### 4. Box Plot

```python
f, ax = plt.pyplot.subplots(1, 1,figsize = (15, 10))
pokemon.boxplot(column = "Sp. Def", by = "Type 1", ax = ax)
```

#### 5. Hexbin Plot

```python
f, ax = plt.pyplot.subplots(1, 1,figsize = (15, 10))
pokemon.plot.hexbin(x = "HP", y = "Defense", ax = ax, gridsize = 8 )
```
